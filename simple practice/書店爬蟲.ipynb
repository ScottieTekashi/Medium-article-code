```python
href_list =['https://books.toscrape.com/'+ i['href'] for i in websourcesoup.select('.product_pod h3 a')]
href_list #串書本網頁
```

```python
list_web = ['https://books.toscrape.com/catalogue/page-{}.html'.format(i) for i in range(1,51)]
list_web #串書品頁數 網址要從第2頁開始
```

```python
href_list=[]
for i in range(0,3):
  res_web = requests.get(list_web[i])
  soup_web = BeautifulSoup(res_web.text)
  for k in soup_web.select('.product_pod h3 a'):
    href ='https://books.toscrape.com/catalogue/'+ k['href']
    href_list.append(href)
href_list 
```

```python
is_UPC = soup.select('.table-striped td')[0].text
print(is_UPC)
```

```python
title=[]
UPC=[]
price=[]
description=[]

for i in href_list:
  res = requests.get(i)
  soup = BeautifulSoup(res.text)
  is_description = soup.select('.product_page p')[3].text
  description.append(is_description)
  is_title = soup.select('.product_main h1')[0].text
  title.append(is_title)
  is_UPC = soup.select('.table-striped td')[0].text
  UPC.append(is_UPC)
  is_price = soup.select('.table-striped td')[2].text
  price.append(is_price)
print(UPC)
```

```python
import pandas as pd

import os
#os.getcwd() 獲取路徑
```

```python
df1 = pd.DataFrame(title, columns = ['title'])
```

```python
df1 = pd.concat([df1, pd.DataFrame(UPC,columns=['UPC']),pd.DataFrame(price, columns = ['price']),pd.DataFrame(description, columns = ['description'])],axis=1)
```

```python
df1.to_csv('book_info', index=True)
```
